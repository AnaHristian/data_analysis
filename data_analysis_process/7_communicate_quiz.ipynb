```python
# imports and load data
import pandas as pd
% matplotlib inline

df = pd.read_csv('store_data.csv')
df.head(20)
```

```python
df.tail(20)
```

```python
# sales for the last month
df.iloc[196:,1:].sum().plot(kind='bar');
```

```python
# average sales
df.iloc[:,1:].mean().plot(kind='bar');
```

```python
# sales for the week of March 13th, 2016
df[df['week'] == '2016-03-13'].iloc[:,1:].plot(kind='bar');
```

```python
df[df['week'] >= '2017-12-01'].iloc[:,1:].sum().plot(kind='pie', figsize=(8,8))
```
